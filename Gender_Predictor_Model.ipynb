```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from pickle import dump
```

```python
gen_class = pd.read_csv('/content/drive/MyDrive/weight-height.csv')
gen_class
```

```python
gen_class['Gender'] = gen_class['Gender'].replace(to_replace=['Male', 'Female'], value=[1, 0])
gen_class
```

```python
gen_class['Height_cm']=gen_class['Height']*2.54
gen_class['Weight_kg']=gen_class['Weight']*2.205
gen_class=gen_class.drop(['Height','Weight'],axis=1)
gen_class
```

```python
x=gen_class.drop(['Gender'],axis=1)
y=gen_class['Gender']
lr_model = LogisticRegression()
lr_model.fit(x,y)
```

```python
lr_model.score(x,y)
# .score(x,y) tells how good your model is at predicting the output.
```

```python
y_pred = lr_model.predict(x)
# .predict(x) predicts the output for the given input!
```

```python
lr_model.predict_proba(x)
# .predict_proba(x)(x) predicts the probability as output for the given input!

# [4.66e-06, 9.99995e-01] -> Probability of Class 0 = 0.00000466 (almost 0%) &
# Probability of Class 1 = 0.999995 (almost 100%)

# [0.724, 0.276] -> Class 0 â†’ 72.4% & Class 1 â†’ 27.6%
# The model chooses Class 0, but confidence is lower.
```

```python
accuracy_score(y, y_pred)
# gives the accuracy of the model!
```

```python
dump(lr_model, open('model.pkl', 'wb'))
```
